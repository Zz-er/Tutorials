## MAP-Elites 解决什么问题？

在第一章中，我们构建了一个简单的进化优化器，并观察到了一个致命缺陷：**种群会收敛**。所有存活的程序都是同一种冒泡排序的微小变体。我们永远不会发现完全不同的方法，比如插入排序或归并排序。

这是经典的**质量 vs. 多样性**困境：

- **只追求质量**（只保留最好的）：会陷入局部最优。如果带微调的冒泡排序得分 0.52，不带微调的得分 0.50，我们永远保留微调版——即使归并排序目前只有 0.45（还没优化），它最终可能达到 0.90。

- **只追求多样性**（什么都保留）：种群爆炸，搜索变成随机游走。

**MAP-Elites** 的解决方案：为**每种类型的解决方案**保留最好的程序。就像组织一个奖杯柜：每个类别一个架子，每个架子只放该类别的冠军。

## 直觉：奖杯柜

想象你在举办一场烹饪比赛。你可以只选出最好的一道菜。但如果所有人都做意大利菜，你永远不会发现有人能做出惊艳的泰国咖喱。

MAP-Elites 就像按照以下方式组织比赛：
- **行：** 菜系类型（意大利、泰国、墨西哥……）
- **列：** 复杂程度（简单、中等、精致）

每个格子保留**该菜系和复杂度下最好的菜品**。你自然获得了多样性（多种菜系都有代表）和质量（每个格子的菜品都是该类别最好的）。

对于代码进化，「特征」可能是：
- **代码长度**（短程序 vs. 长程序）
- **算法复杂度**（O(n²) vs. O(n log n)）

这样我们同时保留了*又短又简单的冒泡排序*和*虽长但快的归并排序*。

## 理论：MAP-Elites 算法

### 形式化定义

设 $\mathcal{X}$ 为所有程序的空间。我们定义：

- $f: \mathcal{X} \to \mathbb{R}$ — **适应度函数**（越高越好，例如速度分数）
- $\phi: \mathcal{X} \to \mathbb{R}^d$ — **特征描述符**，将程序映射到 $d$ 个特征值
- $\mathcal{B} = \{1, \ldots, B\}^d$ — 一个离散**网格**，每个维度有 $B$ 个桶
- $\text{bin}: \mathbb{R}^d \to \mathcal{B}$ — 离散化函数

### 网格

MAP-Elites 网格 $\mathcal{G}$ 是一个将网格单元映射到程序的字典：

$$
\mathcal{G}: \mathcal{B} \to \mathcal{X} \cup \{\emptyset\}
$$

每个单元最多保存一个程序——具有这些特征的最高适应度程序。

### 更新规则

当产生一个新程序 $x'$ 时：

$$
\begin{aligned}
& \textbf{步骤 1:} \quad \mathbf{b} = \text{bin}(\phi(x')) \quad \text{// 计算特征单元} \\
& \textbf{步骤 2:} \quad \textbf{if } \mathcal{G}[\mathbf{b}] = \emptyset \text{ or } f(x') > f(\mathcal{G}[\mathbf{b}]): \\
& \qquad\qquad \mathcal{G}[\mathbf{b}] \leftarrow x' \quad \text{// 如果更好或单元为空则替换}
\end{aligned}
$$

### 优化目标

MAP-Elites 隐式地优化：

$$
\max_{\mathcal{G}} \sum_{\mathbf{b} \in \mathcal{B}} f(\mathcal{G}[\mathbf{b}]) \cdot \mathbb{1}[\mathcal{G}[\mathbf{b}] \neq \emptyset]
$$

用大白话说：最大化所有已占用单元的总适应度。这自然奖励了**高适应度**（每个单元的冠军都很好）和**覆盖率**（更多单元被占用）。

### 完整算法

$$
\begin{aligned}
& \textbf{算法：} \text{ MAP-Elites} \\
& \textbf{输入：} \text{ 初始程序 } x_0, \text{ 特征描述符 } \phi, \text{ 适应度 } f \\
& \mathcal{G} \leftarrow \{\} \quad \text{// 空网格} \\
& \text{Add}(\mathcal{G}, x_0) \quad \text{// 添加初始程序} \\
& \textbf{for } t = 1, 2, \ldots, T: \\
& \quad x \leftarrow \text{从 } \mathcal{G} \text{ 中采样} \quad \text{// 选择父代（在已占用单元上均匀随机）} \\
& \quad x' \leftarrow \text{mutate}(x) \\
& \quad \mathbf{b} \leftarrow \text{bin}(\phi(x')) \\
& \quad \textbf{if } \mathcal{G}[\mathbf{b}] = \emptyset \textbf{ or } f(x') > f(\mathcal{G}[\mathbf{b}]): \\
& \quad \quad \mathcal{G}[\mathbf{b}] \leftarrow x' \\
& \textbf{return } \mathcal{G}
\end{aligned}
$$

> **出处：** Mouret, J.-B. & Clune, J. (2015). *Illuminating search spaces by mapping elites.* [arXiv:1504.04909](https://arxiv.org/abs/1504.04909), Algorithm 1  
> **对应原始代码：** `openevolve/database.py:L400-L500` — `add_program()` 实现了上述更新规则

### 数学 → 代码 映射表

| 数学符号 | 代码变量 | 含义 |
|----------|----------|------|
| $\mathcal{X}$ | `str`（程序代码） | 所有程序的空间 |
| $f(x)$ | `get_fitness(program)` | 适应度分数（越高越好） |
| $\phi(x)$ | `get_features(program)` | 特征描述符（如代码长度、复杂度） |

| $\text{bin}(\cdot)$ | `discretize(features, n_bins)` | 将连续特征映射到网格索引 |
| $\mathcal{G}$ | `self.grid: dict[tuple, Program]` | MAP-Elites 网格 |
| $B$ | `self.n_bins` | 每个维度的桶数 |
| $d$ | `len(self.feature_dims)` | 特征维度数 |

## 实现：MAP-Elites 网格

现在让我们从零实现。我们将构建三个组件：
1. **特征提取** — 计算程序的特征
2. **网格** — MAP-Elites 数据结构
3. **更新后的进化循环** — 将网格集成到第一章的循环中

```python
import sys, os
sys.path.insert(0, os.path.abspath(".."))

import random
import math
from dataclasses import dataclass, field
from typing import Optional

# 复用第一章的 Program 类
from our_implementation.core import Program

print("导入完成")
```

```python
def get_features(program: Program, feature_dims: list[str]) -> list[float]:
    """从程序中提取特征值。
    
    这实现了上面数学公式中的 φ(x)。
    
    支持的特征维度：
    - 'complexity': 代码字符长度（作为算法复杂度的代理指标）
    - 'score': 评估分数
    - program.metrics 中的任意键
    
    设计决策：我们用代码长度作为复杂度的代理指标，因为它容易计算
    且与实际算法复杂度相关。OpenEvolve (database.py:L300) 也支持
    从评估器返回的自定义指标。
    """
    features = []
    for dim in feature_dims:
        if dim == "complexity":
            # 代码长度作为复杂度代理
            features.append(float(len(program.code)))
        elif dim in program.metrics:
            features.append(float(program.metrics[dim]))
        else:
            features.append(0.0)
    return features


def discretize(features: list[float], n_bins: int,
               mins: list[float], maxs: list[float]) -> tuple[int, ...]:
    """将连续特征值映射到网格单元索引。
    
    这实现了数学公式中的 bin(φ(x))。
    
    步骤（对应推导过程）：
    1. 用 min-max 归一化将每个特征缩放到 [0, 1]
    2. 乘以 n_bins 并取整得到整数索引
    3. 截断到有效范围 [0, n_bins-1]
    
    在 OpenEvolve 中：database.py:_scale_feature_value() 支持
    minmax、zscore 和 percentile 三种缩放方式。我们先用 minmax。
    """
    cell = []
    for val, lo, hi in zip(features, mins, maxs):
        if hi - lo < 1e-8:
            # 所有值相同——放在中间桶
            cell.append(n_bins // 2)
        else:
            scaled = (val - lo) / (hi - lo)           # 步骤 1: 缩放到 [0,1]
            bin_idx = int(scaled * n_bins)             # 步骤 2: 离散化
            bin_idx = max(0, min(n_bins - 1, bin_idx)) # 步骤 3: 截断
            cell.append(bin_idx)
    return tuple(cell)


# 快速测试
print("discretize([0.3, 0.7], n_bins=10, mins=[0,0], maxs=[1,1]) =",
      discretize([0.3, 0.7], 10, [0, 0], [1, 1]))
# 预期结果: (3, 7)
```

```python
class MAPElitesGrid:
    """MAP-Elites 质量-多样性网格。
    
    实现了理论部分的网格 G：
    - 每个单元最多保存一个程序（该特征组合下的最佳程序）
    - 添加程序时，只有适应度更高才会替换已有程序
    - 采样时从已占用单元中均匀选择
    
    在 OpenEvolve 中：database.py:ProgramDatabase 为每个岛屿
    管理一个这样的网格。我们将在第三章添加岛屿机制。
    """
    
    def __init__(self, feature_dims: list[str], n_bins: int = 10):
        """
        Args:
            feature_dims: 特征维度名称（如 ['complexity', 'score']）
            n_bins: 每个维度的桶数。总单元数 = n_bins^d
            
        设计决策：默认每个维度 10 个桶。
        OpenEvolve 使用可配置的 feature_bins（默认 10）。
        太少 = 多样性差。太多 = 网格稀疏，覆盖慢。
        """
        self.feature_dims = feature_dims
        self.n_bins = n_bins
        self.grid: dict[tuple[int, ...], Program] = {}
        
        # 自适应特征范围——随着看到更多程序而更新
        # 设计决策：从极端默认值开始，随着观察数据逐步收窄。
        # 这很重要，因为我们事先不知道特征的范围。
        # OpenEvolve (database.py:L300) 也是这样做的。
        self.feature_mins = [float('inf')] * len(feature_dims)
        self.feature_maxs = [float('-inf')] * len(feature_dims)
        
        # 跟踪绝对最佳（永远不会丢失，即使其单元被覆盖）
        self.best: Optional[Program] = None
    
    def _update_ranges(self, features: list[float]):
        """随着看到新的特征值，自适应更新 min/max 范围。"""
        for i, val in enumerate(features):
            self.feature_mins[i] = min(self.feature_mins[i], val)
            self.feature_maxs[i] = max(self.feature_maxs[i], val)
    
    def _get_fitness(self, program: Program) -> float:
        """从程序的指标中提取适应度分数。
        
        在 OpenEvolve 中：get_fitness_score() 对所有数值指标
        取平均（排除特征维度）。我们简化为只取 'score'。
        """
        return program.metrics.get("score", 0.0)
    
    def add(self, program: Program) -> bool:
        """将程序添加到网格。返回 True 表示已存储。
        
        实现理论部分的更新规则：
          if G[b] 为空 or f(x') > f(G[b]): G[b] = x'
        """
        features = get_features(program, self.feature_dims)
        self._update_ranges(features)
        
        cell = discretize(features, self.n_bins,
                          self.feature_mins, self.feature_maxs)
        
        fitness = self._get_fitness(program)
        
        # 更新绝对最佳（OpenEvolve: database.py 单独跟踪）
        if self.best is None or fitness > self._get_fitness(self.best):
            self.best = program
        
        # MAP-Elites 更新规则
        if cell not in self.grid:
            # 单元为空 -> 添加
            self.grid[cell] = program
            return True
        elif fitness > self._get_fitness(self.grid[cell]):
            # 比当前占据者更好 -> 替换
            self.grid[cell] = program
            return True
        else:
            # 更差 -> 丢弃
            return False
    
    def sample(self) -> Program:
        """从已占用的单元中均匀随机采样一个程序。
        
        因为我们是在单元上均匀采样（而不是在程序上），
        稀疏区域的程序被采样的概率更高。
        这是关键的多样性机制。
        """
        if not self.grid:
            raise ValueError("网格为空")
        return random.choice(list(self.grid.values()))
    
    def __len__(self) -> int:
        """已占用的单元数。"""
        return len(self.grid)
    
    @property
    def coverage(self) -> float:
        """已占用单元占总单元数的比例。"""
        total_cells = self.n_bins ** len(self.feature_dims)
        return len(self.grid) / total_cells


print(f"MAPElitesGrid 就绪。2 个维度、10 个桶时：共 {10**2} 个单元")
```

## 演示：观察网格被填充

让我们创建一个网格，添加 200 个随机程序，看看 MAP-Elites 如何维持多样性：

```python
import numpy as np

random.seed(42)
np.random.seed(42)

# 创建 2D 网格：复杂度（代码长度）x 分数
grid = MAPElitesGrid(feature_dims=["complexity", "score"], n_bins=10)

# 生成 200 个合成程序，具有不同的复杂度和分数
stored_count = 0
for i in range(200):
    # 模拟具有随机特征的程序
    code_length = random.randint(50, 500)
    fake_code = "x" * code_length  # 占位代码
    score = random.random() * 0.8 + 0.1  # 分数在 [0.1, 0.9]
    
    program = Program(
        code=fake_code,
        metrics={"score": score, "complexity": float(code_length)},
        generation=i,
    )
    
    if grid.add(program):
        stored_count += 1

print(f"添加了 200 个程序：")
print(f"  已存储: {stored_count}（其余被丢弃——单元中已有更好的程序）")
print(f"  已占用单元: {len(grid)} / {10**2} ({grid.coverage:.0%} 覆盖率)")
print(f"  最佳适应度: {grid.best.metrics['score']:.4f}")
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# 构建网格的热力图
heatmap = np.full((10, 10), np.nan)  # NaN = 空单元
for cell, program in grid.grid.items():
    row, col = cell
    heatmap[row, col] = program.metrics["score"]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 左图：占用情况（哪些单元有程序？）
occupancy = ~np.isnan(heatmap)
axes[0].imshow(occupancy.astype(float), cmap="Greens", aspect="equal",
               origin="lower", vmin=0, vmax=1)
axes[0].set_xlabel("分数桶 →")
axes[0].set_ylabel("复杂度桶 →")
axes[0].set_title(f"网格占用情况 ({grid.coverage:.0%} 已填充)")
for i in range(10):
    for j in range(10):
        text = "●" if occupancy[i, j] else "·"
        axes[0].text(j, i, text, ha="center", va="center",
                     fontsize=12, color="white" if occupancy[i, j] else "gray")

# 右图：适应度热力图（每个已占用单元的分数是多少？）
cmap = plt.cm.viridis.copy()
cmap.set_bad(color="#f0f0f0")  # 灰色表示空单元
im = axes[1].imshow(heatmap, cmap=cmap, aspect="equal", origin="lower",
                     vmin=0, vmax=1)
axes[1].set_xlabel("分数桶 →")
axes[1].set_ylabel("复杂度桶 →")
axes[1].set_title("每个单元的适应度（每个单元中的最佳程序）")
plt.colorbar(im, ax=axes[1], label="适应度分数")

plt.suptitle("MAP-Elites 网格 — 200 个程序，10×10 网格",
             fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()

print("每个已填充的单元只保留该特征组合下最好的程序。")
print("这是核心的多样性机制——与第一章的列表有本质区别。")
```

## 对比：扁平列表 vs. MAP-Elites

让我们直接对比第一章的方法（保留所有程序）和 MAP-Elites：

```python
from our_implementation.core import ProgramStore

random.seed(42)

# 同样的 200 个程序，两种不同的存储方式
flat_store = ProgramStore()
me_grid = MAPElitesGrid(feature_dims=["complexity", "score"], n_bins=10)

for i in range(200):
    code_length = random.randint(50, 500)
    score = random.random() * 0.8 + 0.1
    p = Program(
        code="x" * code_length,
        metrics={"score": score, "complexity": float(code_length)},
        generation=i,
    )
    flat_store.add(p)
    me_grid.add(p)

# 对比：从各自采样 100 个父代，查看多样性
flat_scores = [flat_store.sample_parent().metrics["score"] for _ in range(100)]
me_scores = [me_grid.sample().metrics["score"] for _ in range(100)]

flat_complexity = [len(flat_store.sample_parent().code) for _ in range(100)]
me_complexity = [len(me_grid.sample().code) for _ in range(100)]

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(flat_scores, bins=20, alpha=0.6, label="扁平列表", color="#ef4444")
axes[0].hist(me_scores, bins=20, alpha=0.6, label="MAP-Elites", color="#2563eb")
axes[0].set_xlabel("采样父代的分数")
axes[0].set_ylabel("计数")
axes[0].set_title("采样父代的分数分布")
axes[0].legend()

axes[1].hist(flat_complexity, bins=20, alpha=0.6, label="扁平列表", color="#ef4444")
axes[1].hist(me_complexity, bins=20, alpha=0.6, label="MAP-Elites", color="#2563eb")
axes[1].set_xlabel("采样父代的代码长度")
axes[1].set_ylabel("计数")
axes[1].set_title("采样父代的复杂度分布")
axes[1].legend()

plt.suptitle("采样对比：扁平列表 vs. MAP-Elites", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()

print(f"扁平列表：{len(flat_store)} 个程序已存储（全部保留）")
print(f"MAP-Elites：{len(me_grid)} 个程序已存储（只保留单元冠军）")
print(f"\nMAP-Elites 存储的程序数量减少了 {len(flat_store) / len(me_grid):.0f} 倍，但保持了多样性。")
```

## 深入：自适应特征缩放

一个微妙之处：网格的桶需要覆盖正确的范围。如果代码长度范围是 50 到 500，那么桶 0 应该大约是 50-95，桶 1 大约是 95-140，等等。

但我们事先不知道范围！第一个程序可能长度为 100，第 200 个可能长度为 5000。

**解决方案：自适应缩放。** 我们跟踪每个特征的最小值和最大值，并动态重新缩放。

> **设计决策：** OpenEvolve（`database.py:_scale_feature_value`）支持三种方法：
> - **MinMax**（我们实现的）— 简单，实践中效果好
> - **Z-score** — 更好地处理异常值
> - **Percentile** — 对偏态分布最鲁棒
>
> 我们先用 MinMax，因为它最容易理解。重要的是缩放是*自适应的*——它随着我们看到更多数据而变化。

我们的 `_update_ranges()` 方法已经处理了这个问题。以下展示当新的极值到来时会发生什么：

```python
# 演示自适应缩放
demo_grid = MAPElitesGrid(feature_dims=["complexity"], n_bins=5)

print("添加代码长度递增的程序...")
print(f"{'程序':>10} {'长度':>8} {'最小值':>10} {'最大值':>10} {'单元':>6}")
print("-" * 50)

for length in [100, 200, 50, 500, 300]:
    p = Program(code="x" * length, metrics={"score": 0.5, "complexity": float(length)})
    demo_grid.add(p)
    features = get_features(p, ["complexity"])
    cell = discretize(features, 5, demo_grid.feature_mins, demo_grid.feature_maxs)
    print(f"{'x*'+str(length):>10} {length:>8} {demo_grid.feature_mins[0]:>10.0f} "
          f"{demo_grid.feature_maxs[0]:>10.0f} {cell[0]:>6}")

print("\n注意 min/max 范围如何随着新程序的到来而自适应调整。")
print("早期程序可能因为范围扩展而切换单元——这是预期行为。")
```

## 对应 OpenEvolve 源码

下表展示了我们的实现与 OpenEvolve 源码的对应关系：

| 我们的实现 | OpenEvolve 源码 | 说明 |
|-----------|----------------|------|
| `get_features()` | `database.py:_extract_features()` | 特征提取，OpenEvolve 支持更多自定义特征 |
| `discretize()` | `database.py:_scale_feature_value()` | 离散化，OpenEvolve 额外支持 zscore 和 percentile |
| `MAPElitesGrid` | `database.py:ProgramDatabase` | 网格管理，OpenEvolve 每个岛屿一个网格 |
| `grid.add()` | `database.py:add_program()` | 更新规则完全一致 |
| `grid.sample()` | `database.py:sample_from_island()` | 采样，OpenEvolve 有更复杂的选择策略 |
| `grid.best` | `database.py:get_best_program()` | 绝对最佳跟踪，逻辑相同 |

## 保存到 `our-implementation/`

把我们的 MAP-Elites 实现保存为一个模块：

```python
MODULE_CODE = '''
"""our-implementation/map_elites.py — MAP-Elites 质量-多样性网格。

在第二章中构建。实现了以下论文的算法：
Mouret & Clune (2015), "Illuminating search spaces by mapping elites"
"""
import random
from typing import Optional
from .core import Program


def get_features(program: Program, feature_dims: list[str]) -> list[float]:
    """从程序中提取特征值（实现 phi(x)）。"""
    features = []
    for dim in feature_dims:
        if dim == "complexity":
            features.append(float(len(program.code)))
        elif dim in program.metrics:
            features.append(float(program.metrics[dim]))
        else:
            features.append(0.0)
    return features


def discretize(features: list[float], n_bins: int,
               mins: list[float], maxs: list[float]) -> tuple[int, ...]:
    """将连续特征映射到网格单元索引（实现 bin(phi(x))）。"""
    cell = []
    for val, lo, hi in zip(features, mins, maxs):
        if hi - lo < 1e-8:
            cell.append(n_bins // 2)
        else:
            scaled = (val - lo) / (hi - lo)
            bin_idx = max(0, min(n_bins - 1, int(scaled * n_bins)))
            cell.append(bin_idx)
    return tuple(cell)


class MAPElitesGrid:
    """MAP-Elites 质量-多样性网格。"""

    def __init__(self, feature_dims: list[str], n_bins: int = 10):
        self.feature_dims = feature_dims
        self.n_bins = n_bins
        self.grid: dict[tuple[int, ...], Program] = {}
        self.feature_mins = [float("inf")] * len(feature_dims)
        self.feature_maxs = [float("-inf")] * len(feature_dims)
        self.best: Optional[Program] = None

    def _update_ranges(self, features: list[float]):
        for i, val in enumerate(features):
            self.feature_mins[i] = min(self.feature_mins[i], val)
            self.feature_maxs[i] = max(self.feature_maxs[i], val)

    def _get_fitness(self, program: Program) -> float:
        return program.metrics.get("score", 0.0)

    def add(self, program: Program) -> bool:
        features = get_features(program, self.feature_dims)
        self._update_ranges(features)
        cell = discretize(features, self.n_bins,
                          self.feature_mins, self.feature_maxs)
        fitness = self._get_fitness(program)
        if self.best is None or fitness > self._get_fitness(self.best):
            self.best = program
        if cell not in self.grid or fitness > self._get_fitness(self.grid[cell]):
            self.grid[cell] = program
            return True
        return False

    def sample(self) -> Program:
        if not self.grid:
            raise ValueError("网格为空")
        return random.choice(list(self.grid.values()))

    def __len__(self) -> int:
        return len(self.grid)

    @property
    def coverage(self) -> float:
        return len(self.grid) / (self.n_bins ** len(self.feature_dims))
'''

with open("../our-implementation/map_elites.py", "w", encoding="utf-8") as f:
    f.write(MODULE_CODE)

print("已保存 our-implementation/map_elites.py")
print("  导出: get_features, discretize, MAPElitesGrid")
```

## 本章小结

我们构建了：

- **`get_features()`** — 从程序中提取特征值（实现 $\phi(x)$）
- **`discretize()`** — 将特征映射到网格单元索引（实现 $\text{bin}(\cdot)$）
- **`MAPElitesGrid`** — 质量-多样性网格（实现 $\mathcal{G}$）

### 核心要点

1. **MAP-Elites 每个单元保留一个冠军** — 有界内存，保证多样性
2. **从单元（而非程序）中采样** 给予稀疏区域同等权重
3. **自适应特征缩放** 处理未知的值域范围
4. **绝对最佳单独跟踪** — 永远不会因为单元替换而丢失

### 还缺什么？

只有一个全局网格时，所有程序竞争同样的单元。如果一种有前途但尚未优化的方法出现，它可能被当前最佳的精调变体覆盖。我们需要**隔离**——多个网格独立进化。
