## 问题：优化代码太难了

假设你有一个排序函数，它能跑——但它是**最好的**排序方式吗？换一种算法会不会更快？改几个常量、换一种数据结构、甚至完全换一种思路，效果会不会更好？

传统做法需要人类专家：懂文献、看得懂 profiling 数据、还能写出（并测试）更好代码的那种专家。这种专家贵，而且慢。

**如果能自动化呢？** 把一段代码丢给 AI，给它一个衡量“更好”的标准，让它自己去探索改进空间——行不行？

这正是 OpenEvolve 做的事。它是 Google DeepMind [AlphaEvolve](https://deepmind.google/discover/blog/alphaevolve-a-gemini-powered-coding-agent-for-designing-advanced-algorithms/) 的开源实现——一个用 LLM（大语言模型）当“变异算子”的进化算法系统。

### 原版 AlphaEvolve 的真实成果

- 发现了一种 **4×4 矩阵乘法** 算法，只用 48 次乘法（已知最优是 49 次）
- 找到了在某些尺寸上**超越人类设计**的排序网络
- 将 **GPU 核函数**性能优化了 20-30%

同样的方法，开源在 OpenEvolve 中，已经在 MLX Metal 核函数上实现了 **2.8 倍加速**。

## 核心洞察：进化 + LLM

经典进化算法通过随机位翻转或交换来变异代码——它们不理解代码的语义，所以很慢。

LLM 则能理解代码、解释改动、生成*有意义*的修改。但它们不会系统性地搜索——只是给你一个答案。

**OpenEvolve 把两者结合了：**

1. **LLM** 充当*智能变异算子*——它读懂已有代码，理解上下文，生成改进方案
2. **进化算法** 提供*搜索策略*——维护种群、选择父代、跟踪进展、避免陷入局部最优

可以这样理解：**LLM 提供创造力，进化算法提供纪律性。**

## 架构全景

从一万英尺高空俯瞰，系统是这样运作的：

```mermaid
flowchart TD
    User(["👤 用户\n（初始程序 + 评估器）"])
    Controller["🎛️ 控制器 Controller\n统筹全局"]
    DB[("🗄️ 数据库 Database\nMAP-Elites + 岛屿")]
    Eval["⚖️ 评估器 Evaluator\n级联打分"]
    LLM["🤖 LLM 集成\n代码生成"]
    Workers["⚙️ 工作进程\n并行迭代"]

    User -->|初始代码| Controller
    Controller --> DB
    Controller --> Workers
    Workers -->|采样父代| DB
    Workers -->|生成变异| LLM
    Workers -->|评分| Eval
    Workers -->|存储结果| DB
    DB -->|最优程序| Controller
    Controller -->|结果| User
```

核心组件一览：

| 组件 | 文件 | 职责 |
|------|------|------|
| `Controller` 控制器 | `controller.py` | 顶层循环、检查点、并行调度 |
| `ProgramDatabase` 程序数据库 | `database.py` | 存储所有程序、MAP-Elites 网格、岛屿 |
| `Evaluator` 评估器 | `evaluator.py` | 运行程序、三阶段质量打分 |
| `LLMEnsemble` LLM 集成 | `llm/ensemble.py` | 调用一个或多个 LLM 生成代码改动 |
| `PromptSampler` 提示词采样器 | `prompt/sampler.py` | 构建展示给 LLM 的提示词 |
| `run_iteration` 单次迭代 | `iteration.py` | 一个完整的 采样→生成→评估→存储 周期 |

## 我们将一起构建什么

我们会按**认知顺序**从零实现 OpenEvolve——也就是说，每个概念只在你**需要**它的时候才引入，而且一定是在你**感受到了它要解决的痛点**之后。

### 贯穿全文的例子

整个教程系列中，我们会持续演化同一个例子：**一个排序算法**。

- **输入：** 一个 Python 排序函数（最初是冒泡排序）
- **评估器：** 在 1000 个随机元素上测量速度
- **目标：** 通过 LLM 驱动的进化，发现更快的排序算法

选择排序作为例子是因为：排序人人都懂、评估客观且快、解空间丰富（插入排序、归并排序、Timsort……）。
