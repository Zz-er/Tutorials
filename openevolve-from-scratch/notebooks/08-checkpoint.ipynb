## 1. 痛点：进度丢失

上一章我们实现了并行进化，可以同时跑多个 Worker。但想象这个场景：

```
进化实验 — AlgoTune 优化任务
  迭代 1-500:   ████████████████████ 正常运行中...
  迭代 501:     💥 断电 / OOM / 网络中断

结果：8 小时的进化成果全部丢失，从头来过。
```

更常见的场景：
- GPU 机器被人抢占
- API 调用额度用完，想明天继续
- 想调整参数（比如增加迭代数），但不想从头跑

**解决方案：** 定期保存完整的进化状态（检查点），崩溃后从最近的检查点恢复。

**生活类比：** 就像游戏的存档——你不会每次开机都从第一关开始打。

## 2. 理论：需要保存什么？

一个完整的进化状态包含：

```mermaid
flowchart TB
    subgraph "检查点内容"
        META[metadata.json<br/>迭代计数、岛屿状态、特征缩放]
        PROGS[programs/<br/>每个程序一个 JSON 文件]
        BEST[best_program.py<br/>当前最优代码]
        INFO[best_program_info.json<br/>最优程序的指标]
        ART[artifacts/<br/>大型工件文件]
    end
```

### 2.1 检查点目录结构

```
openevolve_output/
├── checkpoints/
│   ├── checkpoint_100/       # 第 100 次迭代的快照
│   │   ├── metadata.json     # 元数据
│   │   ├── programs/         # 所有程序
│   │   │   ├── abc123.json
│   │   │   ├── def456.json
│   │   │   └── ...
│   │   ├── best_program.py   # 最优代码
│   │   └── best_program_info.json
│   ├── checkpoint_200/       # 第 200 次迭代的快照
│   └── ...
└── best/                     # 最终最优结果
```

### 2.2 保存频率

$$\text{checkpoint\_cost} = N_{programs} \times S_{avg} + S_{metadata}$$

**数值例子：**
> 500 个程序，每个 JSON 约 2KB，metadata 约 10KB  
> $\text{cost} = 500 \times 2\text{KB} + 10\text{KB} = 1010\text{KB} \approx 1\text{MB}$  
> 写入 SSD 约 10ms——完全可以接受！

OpenEvolve 默认每 **100 次迭代**保存一次检查点（`checkpoint_interval: 100`）。

**生活类比：** 就像写论文时按 Ctrl+S——太频繁浪费时间，太少了可能丢进度。  
每 100 次迭代保存一次，是性能和安全之间的好平衡。

## 3. 从零实现：程序序列化

首先，每个程序需要能转换成 JSON 格式（序列化）和从 JSON 恢复（反序列化）。

````python
import json
import os
import time
from dataclasses import dataclass, field, asdict
from typing import Dict, Optional, Any, List


@dataclass
class Program:
    """可序列化的程序对象"""
    id: str
    code: str
    metrics: Dict[str, float] = field(default_factory=dict)
    parent_id: Optional[str] = None
    generation: int = 0
    island: int = 0
    timestamp: float = field(default_factory=time.time)
    iteration_found: int = 0
    metadata: Dict[str, Any] = field(default_factory=dict)
    
    def to_dict(self) -> dict:
        """序列化为字典（可直接 JSON 化）"""
        return asdict(self)
    
    @classmethod
    def from_dict(cls, d: dict) -> 'Program':
        """从字典反序列化"""
        return cls(**{k: v for k, v in d.items() if k in cls.__dataclass_fields__})
    
    def save(self, directory: str):
        """保存到文件"""
        os.makedirs(directory, exist_ok=True)
        path = os.path.join(directory, f"{self.id}.json")
        with open(path, 'w', encoding='utf-8') as f:
            json.dump(self.to_dict(), f, indent=2, ensure_ascii=False)
    
    @classmethod
    def load(cls, path: str) -> 'Program':
        """从文件加载"""
        with open(path, 'r', encoding='utf-8') as f:
            return cls.from_dict(json.load(f))


# 测试序列化/反序列化
prog = Program(
    id='test_001',
    code='def sort_func(arr): return sorted(arr)',
    metrics={'combined_score': 0.95, 'speed': 0.8},
    parent_id='parent_000',
    generation=5,
    island=2,
    iteration_found=42,
)

# 序列化
d = prog.to_dict()
print("序列化结果:")
print(json.dumps(d, indent=2, ensure_ascii=False)[:300] + '...')

# 反序列化
restored = Program.from_dict(d)
print(f"\n反序列化: id={restored.id}, score={restored.metrics['combined_score']}")
print(f"代码一致: {restored.code == prog.code}")
````

## 4. 从零实现：检查点管理器

检查点管理器负责：
1. **保存**：将完整进化状态写入磁盘
2. **加载**：从磁盘恢复进化状态
3. **目录管理**：创建和组织检查点目录

### 设计决策

| 问题 | OpenEvolve 的选择 | 原因 |
|------|-------------------|------|
| 序列化格式 | JSON | 人类可读、跨语言、调试方便 |
| 程序存储 | 每个程序一个文件 | 避免单文件过大、方便增量更新 |
| 元数据 | 单个 metadata.json | 全局状态集中管理 |
| 最优程序 | 单独保存代码文件 | 方便直接查看和使用 |

````python
import shutil
import glob


class CheckpointManager:
    """检查点管理器：保存和恢复进化状态"""
    
    def __init__(self, output_dir: str, checkpoint_interval: int = 100):
        self.output_dir = output_dir
        self.checkpoint_interval = checkpoint_interval
        self.checkpoints_dir = os.path.join(output_dir, 'checkpoints')
        os.makedirs(self.checkpoints_dir, exist_ok=True)
    
    def should_save(self, iteration: int) -> bool:
        """判断是否应该保存检查点"""
        return iteration > 0 and iteration % self.checkpoint_interval == 0
    
    def save(self, iteration: int, programs: Dict[str, Program],
             islands: List[List[str]], best_id: Optional[str] = None,
             extra_metadata: dict = None):
        """保存检查点
        
        Args:
            iteration: 当前迭代数
            programs: 所有程序 {id: Program}
            islands: 岛屿分配 [[prog_ids], ...]
            best_id: 最优程序的 ID
            extra_metadata: 额外的元数据
        """
        ckpt_dir = os.path.join(self.checkpoints_dir, f'checkpoint_{iteration}')
        programs_dir = os.path.join(ckpt_dir, 'programs')
        os.makedirs(programs_dir, exist_ok=True)
        
        # 1. 保存所有程序
        for prog_id, prog in programs.items():
            prog.save(programs_dir)
        
        # 2. 保存元数据
        metadata = {
            'last_iteration': iteration,
            'num_programs': len(programs),
            'islands': islands,
            'best_program_id': best_id,
            'timestamp': time.time(),
        }
        if extra_metadata:
            metadata.update(extra_metadata)
        
        meta_path = os.path.join(ckpt_dir, 'metadata.json')
        with open(meta_path, 'w', encoding='utf-8') as f:
            json.dump(metadata, f, indent=2, ensure_ascii=False)
        
        # 3. 保存最优程序代码（方便直接查看）
        if best_id and best_id in programs:
            best_prog = programs[best_id]
            best_path = os.path.join(ckpt_dir, 'best_program.py')
            with open(best_path, 'w', encoding='utf-8') as f:
                f.write(best_prog.code)
            
            info_path = os.path.join(ckpt_dir, 'best_program_info.json')
            with open(info_path, 'w', encoding='utf-8') as f:
                json.dump({
                    'id': best_prog.id,
                    'metrics': best_prog.metrics,
                    'iteration_found': best_prog.iteration_found,
                    'generation': best_prog.generation,
                }, f, indent=2)
        
        return ckpt_dir
    
    def load(self, checkpoint_path: str) -> dict:
        """从检查点加载进化状态
        
        Returns:
            {
                'programs': {id: Program, ...},
                'metadata': {...},
            }
        """
        # 1. 加载元数据
        meta_path = os.path.join(checkpoint_path, 'metadata.json')
        with open(meta_path, 'r', encoding='utf-8') as f:
            metadata = json.load(f)
        
        # 2. 加载所有程序
        programs = {}
        programs_dir = os.path.join(checkpoint_path, 'programs')
        if os.path.exists(programs_dir):
            for fname in os.listdir(programs_dir):
                if fname.endswith('.json'):
                    path = os.path.join(programs_dir, fname)
                    prog = Program.load(path)
                    programs[prog.id] = prog
        
        return {
            'programs': programs,
            'metadata': metadata,
        }
    
    def list_checkpoints(self) -> List[str]:
        """列出所有检查点"""
        pattern = os.path.join(self.checkpoints_dir, 'checkpoint_*')
        dirs = sorted(glob.glob(pattern))
        return dirs
    
    def get_latest(self) -> Optional[str]:
        """获取最新的检查点路径"""
        ckpts = self.list_checkpoints()
        return ckpts[-1] if ckpts else None


print("CheckpointManager 定义完成 ✓")
````

## 5. 演示：保存检查点

模拟一个进化过程，在第 10 次和第 20 次迭代时保存检查点。

````python
import random
import tempfile

# 使用临时目录
tmp_dir = tempfile.mkdtemp(prefix='oe_tutorial_')
ckpt_mgr = CheckpointManager(output_dir=tmp_dir, checkpoint_interval=10)

# 模拟进化过程
programs = {}
islands = [[], [], []]  # 3 个岛屿
best_id = None
best_score = 0.0

sort_variants = [
    ('bubble', 'def sort_func(arr):\n    arr = arr[:]\n    for i in range(len(arr)):\n        for j in range(len(arr)-i-1):\n            if arr[j] > arr[j+1]:\n                arr[j], arr[j+1] = arr[j+1], arr[j]\n    return arr'),
    ('selection', 'def sort_func(arr):\n    arr = arr[:]\n    for i in range(len(arr)):\n        m = i\n        for j in range(i+1, len(arr)):\n            if arr[j] < arr[m]: m = j\n        arr[i], arr[m] = arr[m], arr[i]\n    return arr'),
    ('quick', 'def sort_func(arr):\n    if len(arr) <= 1: return arr\n    p = arr[0]\n    return sort_func([x for x in arr[1:] if x<=p]) + [p] + sort_func([x for x in arr[1:] if x>p])'),
    ('insert', 'def sort_func(arr):\n    arr = arr[:]\n    for i in range(1, len(arr)):\n        key = arr[i]\n        j = i-1\n        while j >= 0 and arr[j] > key:\n            arr[j+1] = arr[j]\n            j -= 1\n        arr[j+1] = key\n    return arr'),
]

random.seed(42)
for i in range(25):
    name, code = random.choice(sort_variants)
    score = random.uniform(0.5, 1.0)
    island = i % 3
    prog_id = f'prog_{i:03d}'
    
    prog = Program(
        id=prog_id,
        code=code,
        metrics={'combined_score': round(score, 3), 'algorithm': name},
        parent_id=f'prog_{max(0,i-3):03d}' if i > 0 else None,
        generation=i // 3,
        island=island,
        iteration_found=i,
    )
    programs[prog_id] = prog
    islands[island].append(prog_id)
    
    if score > best_score:
        best_score = score
        best_id = prog_id
    
    # 检查是否应该保存
    if ckpt_mgr.should_save(i):
        ckpt_dir = ckpt_mgr.save(
            iteration=i,
            programs=programs,
            islands=[list(isl) for isl in islands],
            best_id=best_id,
        )
        print(f"迭代 {i}: 保存检查点到 {os.path.basename(ckpt_dir)}")
        print(f"  程序数: {len(programs)}, 最优分数: {best_score:.3f}")

# 列出所有检查点
ckpts = ckpt_mgr.list_checkpoints()
print(f"\n共 {len(ckpts)} 个检查点:")
for c in ckpts:
    print(f"  {os.path.basename(c)}")
````

## 7. 完整的断点续跑流程

将检查点集成到进化循环中。关键逻辑：
- 启动时检查是否有检查点 → 有则加载，否则从零开始
- 运行中定期保存检查点
- 确保恢复后不重复添加初始程序

````python
def evolution_with_checkpoints(
    initial_code: str,
    mutate_fn,
    evaluate_fn,
    max_iterations: int = 30,
    checkpoint_interval: int = 10,
    output_dir: str = None,
    resume_from: str = None,  # 检查点路径
):
    """带检查点的进化循环"""
    if output_dir is None:
        output_dir = tempfile.mkdtemp(prefix='evo_')
    
    ckpt_mgr = CheckpointManager(output_dir, checkpoint_interval)
    
    # === 初始化或恢复 ===
    if resume_from and os.path.exists(resume_from):
        # 断点续跑
        state = ckpt_mgr.load(resume_from)
        programs = state['programs']
        islands = state['metadata']['islands']
        start_iter = state['metadata']['last_iteration'] + 1
        best_id = state['metadata']['best_program_id']
        best_score = programs[best_id].metrics.get('combined_score', 0) if best_id else 0
        print(f"从检查点恢复: 迭代 {start_iter}, 程序数 {len(programs)}, 最优 {best_score:.3f}")
    else:
        # 从零开始
        programs = {}
        islands = [[], [], []]
        start_iter = 0
        best_id = None
        best_score = 0.0
        
        # 添加初始程序
        init_id = 'initial'
        init_metrics = evaluate_fn(initial_code)
        programs[init_id] = Program(
            id=init_id, code=initial_code,
            metrics=init_metrics, island=0, iteration_found=0,
        )
        islands[0].append(init_id)
        best_id = init_id
        best_score = init_metrics.get('combined_score', 0)
        print(f"从零开始: 初始分数 {best_score:.3f}")
    
    # === 进化循环 ===
    history = {'scores': [], 'checkpoints': []}
    
    for iteration in range(start_iter, max_iterations):
        # 选择父程序
        parent_id = random.choice(list(programs.keys()))
        parent = programs[parent_id]
        
        # 变异
        child_code = mutate_fn(parent.code)
        child_metrics = evaluate_fn(child_code)
        child_score = child_metrics.get('combined_score', 0)
        
        # 存储
        child_id = f'prog_{iteration:04d}'
        island = iteration % len(islands)
        programs[child_id] = Program(
            id=child_id, code=child_code, metrics=child_metrics,
            parent_id=parent_id, generation=parent.generation+1,
            island=island, iteration_found=iteration,
        )
        islands[island].append(child_id)
        
        if child_score > best_score:
            best_score = child_score
            best_id = child_id
        
        history['scores'].append(best_score)
        
        # 保存检查点
        if ckpt_mgr.should_save(iteration):
            ckpt_dir = ckpt_mgr.save(
                iteration=iteration,
                programs=programs,
                islands=[list(i) for i in islands],
                best_id=best_id,
            )
            history['checkpoints'].append(iteration)
            print(f"  [Checkpoint] 迭代 {iteration}: 保存, 最优={best_score:.3f}")
    
    return programs, best_id, history, output_dir


# 评估和变异函数
def eval_sort(code):
    try:
        ns = {}
        exec(code, ns)
        f = ns.get('sort_func')
        if not f: return {'combined_score': 0.0}
        tests = [([3,1,2],[1,2,3]),([],[]),([5,4,3,2,1],[1,2,3,4,5]),([1],[1])]
        passed = sum(1 for i,e in tests if f(i[:]) == e)
        return {'combined_score': passed / len(tests)}
    except:
        return {'combined_score': 0.0}

def mutate_sort(code):
    variants = [
        'def sort_func(arr): return sorted(arr)',
        'def sort_func(arr): return arr',
        'def sort_func(arr):\n    arr=arr[:]\n    arr.sort()\n    return arr',
        code,  # 不变异
    ]
    return random.choice(variants)


# === 第一次运行：从零开始，跑到第 15 次迭代时模拟中断 ===
random.seed(42)
print("=== 第一次运行 ===")
progs1, best1, hist1, out_dir = evolution_with_checkpoints(
    initial_code='def sort_func(arr): return sorted(arr)',
    mutate_fn=mutate_sort,
    evaluate_fn=eval_sort,
    max_iterations=15,
    checkpoint_interval=10,
)
print(f"第一次运行结束: {len(progs1)} 个程序, 最优={progs1[best1].metrics['combined_score']:.3f}")
````

````python
# === 第二次运行：从检查点恢复，继续到第 30 次迭代 ===
print("\n=== 第二次运行（从检查点恢复）===")

# 找到最新检查点
ckpt_mgr2 = CheckpointManager(out_dir, checkpoint_interval=10)
latest_ckpt = ckpt_mgr2.get_latest()
print(f"恢复检查点: {os.path.basename(latest_ckpt)}")

random.seed(123)  # 不同的种子
progs2, best2, hist2, _ = evolution_with_checkpoints(
    initial_code='def sort_func(arr): return sorted(arr)',  # 不会被使用（因为从检查点恢复）
    mutate_fn=mutate_sort,
    evaluate_fn=eval_sort,
    max_iterations=30,
    checkpoint_interval=10,
    output_dir=out_dir,
    resume_from=latest_ckpt,
)
print(f"第二次运行结束: {len(progs2)} 个程序, 最优={progs2[best2].metrics['combined_score']:.3f}")

# 列出所有检查点
all_ckpts = ckpt_mgr2.list_checkpoints()
print(f"\n所有检查点: {[os.path.basename(c) for c in all_ckpts]}")
````

## 8. 可视化：进化树

检查点数据可以用来重建**进化树**——展示程序之间的父子关系。
OpenEvolve 自带一个 Web 可视化工具（`scripts/visualizer.py`），
这里我们用 matplotlib 实现一个简化版本。

````python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = ['SimHei', 'DejaVu Sans']
matplotlib.rcParams['axes.unicode_minus'] = False

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# 图1：适应度曲线 + 检查点标记
ax1 = axes[0]
all_scores = hist1['scores'] + hist2['scores']
ax1.plot(range(len(all_scores)), all_scores, 'b-', linewidth=2, label='Best Score')

# 标记检查点位置
for cp in hist1['checkpoints'] + hist2['checkpoints']:
    idx = cp if cp < len(all_scores) else len(all_scores) - 1
    ax1.axvline(x=idx, color='red', linestyle='--', alpha=0.5)
    ax1.text(idx, 0.05, f'CP{cp}', rotation=90, fontsize=8, color='red')

# 标记崩溃和恢复点
crash_iter = 14  # 第一次运行到 15 次
ax1.axvline(x=crash_iter, color='orange', linewidth=3, alpha=0.7)
ax1.text(crash_iter + 0.5, 0.5, 'Resume', fontsize=11, color='orange', fontweight='bold')

ax1.set_xlabel('Iteration', fontsize=12)
ax1.set_ylabel('Best Score', fontsize=12)
ax1.set_title('Evolution with Checkpoints', fontsize=14)
ax1.set_ylim(-0.05, 1.1)
ax1.grid(True, alpha=0.3)
ax1.legend(fontsize=11)

# 图2：进化树（简化版）
ax2 = axes[1]
# 取最后的程序集构建树
prog_list = sorted(progs2.values(), key=lambda p: p.iteration_found)[:20]

# 布局：x=迭代, y=岛屿
island_colors = {0: '#e74c3c', 1: '#3498db', 2: '#2ecc71'}

for prog in prog_list:
    x = prog.iteration_found
    y = prog.island + random.uniform(-0.1, 0.1)  # 抖动避免重叠
    color = island_colors.get(prog.island, 'gray')
    size = prog.metrics.get('combined_score', 0) * 200 + 20
    
    ax2.scatter(x, y, s=size, c=color, alpha=0.7, edgecolors='black', linewidth=0.5)
    
    # 画父子连线
    if prog.parent_id and prog.parent_id in progs2:
        parent = progs2[prog.parent_id]
        px, py = parent.iteration_found, parent.island
        ax2.annotate('', xy=(x, y), xytext=(px, py),
                     arrowprops=dict(arrowstyle='->', color='gray', alpha=0.3, lw=0.8))

ax2.set_xlabel('Iteration', fontsize=12)
ax2.set_ylabel('Island', fontsize=12)
ax2.set_title('Evolution Tree (size = score)', fontsize=14)
ax2.set_yticks([0, 1, 2])
ax2.set_yticklabels(['Island 0', 'Island 1', 'Island 2'])

from matplotlib.patches import Patch
legend = [Patch(facecolor=c, label=f'Island {i}') for i, c in island_colors.items()]
ax2.legend(handles=legend, loc='upper left')

plt.tight_layout()
plt.show()

# 清理临时目录
shutil.rmtree(tmp_dir, ignore_errors=True)
shutil.rmtree(out_dir, ignore_errors=True)
````

## 11. 保存模块

将本章实现保存到 `our-implementation/` 目录。

````python
save_code = '''
"""检查点管理器 - 第八章实现

保存和恢复进化状态，支持断点续跑。
"""
import json
import os
import time
import glob
from dataclasses import dataclass, field, asdict
from typing import Dict, Optional, Any, List


@dataclass
class Program:
    id: str
    code: str
    metrics: Dict[str, float] = field(default_factory=dict)
    parent_id: Optional[str] = None
    generation: int = 0
    island: int = 0
    timestamp: float = field(default_factory=time.time)
    iteration_found: int = 0
    metadata: Dict[str, Any] = field(default_factory=dict)
    
    def to_dict(self):
        return asdict(self)
    
    @classmethod
    def from_dict(cls, d):
        return cls(**{k: v for k, v in d.items() if k in cls.__dataclass_fields__})
    
    def save(self, directory):
        os.makedirs(directory, exist_ok=True)
        with open(os.path.join(directory, f"{self.id}.json"), "w") as f:
            json.dump(self.to_dict(), f, indent=2)
    
    @classmethod
    def load(cls, path):
        with open(path) as f:
            return cls.from_dict(json.load(f))


class CheckpointManager:
    def __init__(self, output_dir, checkpoint_interval=100):
        self.output_dir = output_dir
        self.checkpoint_interval = checkpoint_interval
        self.checkpoints_dir = os.path.join(output_dir, "checkpoints")
        os.makedirs(self.checkpoints_dir, exist_ok=True)
    
    def should_save(self, iteration):
        return iteration > 0 and iteration % self.checkpoint_interval == 0
    
    def save(self, iteration, programs, islands, best_id=None, extra_metadata=None):
        ckpt_dir = os.path.join(self.checkpoints_dir, f"checkpoint_{iteration}")
        programs_dir = os.path.join(ckpt_dir, "programs")
        os.makedirs(programs_dir, exist_ok=True)
        for prog in programs.values():
            prog.save(programs_dir)
        metadata = {"last_iteration": iteration, "num_programs": len(programs),
                    "islands": islands, "best_program_id": best_id, "timestamp": time.time()}
        if extra_metadata:
            metadata.update(extra_metadata)
        with open(os.path.join(ckpt_dir, "metadata.json"), "w") as f:
            json.dump(metadata, f, indent=2)
        if best_id and best_id in programs:
            with open(os.path.join(ckpt_dir, "best_program.py"), "w") as f:
                f.write(programs[best_id].code)
        return ckpt_dir
    
    def load(self, checkpoint_path):
        with open(os.path.join(checkpoint_path, "metadata.json")) as f:
            metadata = json.load(f)
        programs = {}
        pdir = os.path.join(checkpoint_path, "programs")
        if os.path.exists(pdir):
            for fname in os.listdir(pdir):
                if fname.endswith(".json"):
                    p = Program.load(os.path.join(pdir, fname))
                    programs[p.id] = p
        return {"programs": programs, "metadata": metadata}
    
    def list_checkpoints(self):
        return sorted(glob.glob(os.path.join(self.checkpoints_dir, "checkpoint_*")))
    
    def get_latest(self):
        ckpts = self.list_checkpoints()
        return ckpts[-1] if ckpts else None
'''

import os
os.makedirs('../our-implementation', exist_ok=True)
with open('../our-implementation/checkpoint.py', 'w', encoding='utf-8') as f:
    f.write(save_code)

print('已保存到 our-implementation/checkpoint.py ✓')
````
